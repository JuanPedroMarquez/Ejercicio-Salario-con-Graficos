## <center>Plotly Open Source Graphing Library</center>

El ecosistema Plotly es una serie de herramientas para analizar y visualizar datos de forma interactiva. Tiene una rama de productos y bibliotecas ampliamente usados en la industria y comprende una variedad de lenguajes de programación como js, python, r, julia, matlab, f ect

## LA COMPOSICIÓN PROFESIONAL SE COMPONE DE 3 PASOS:

En Plotly, una "figura" es el objeto central que contiene todos los elementos de una visualización: datos, diseño y configuraciones de interactividad.

Una figura en Plotly se construye usando la clase plotly.graph_objects.Figure(). 
A este objeto de figura se le pueden agregar trazados (como gráficos de líneas, de barras, de dispersión, etc.) y se le pueden especificar opciones de diseño (como títulos, ejes, leyendas, etc.).

## 1

```python
# Primero empezamos creando una figura vacía 
figura = go.Figure()
```

```python
figura # Podemos comprobar como tenemos el lienzo donde vamos a añadir información y gráficos.
```

## 2

```python
# Sobre el lienzo en blanco, añadimos los trazados, por ejemplo : 
figura.add_trace(go.Scatter(x=["hola", "plot", "ly"], y=[1, 3, 2], name="Definición de trazado uno"))
figura.add_trace(go.Bar(x=["hola", "plot", "ly"], y=[1, 3, 2], name="Definición de trazado dos"))
figura.add_trace(go.Scatter(x=["hola", "plot", "ly"], y=[2, 3, 1], name="Definición de trazado tres"))
```

## 3

```python
# El siguiente paso es especificar las opciones de nuestro diseño, actualizando el layout 
figura.update_layout(title="Mi primer Gráfico con Plotly", xaxis_title="Eje X", yaxis_title="Eje Y") 
#y mostrar la figura con la información añadida
figura.show()
```

Una vez que tenemos definido los 3 pasos, creación de lienzo, añadido de gráfica y actualización con información, podemos ver varios ejemplos de como usar este abanico de gráficos

```python
#Carga de datos de prueba 
df = px.data.iris()
```

## Gráficos mas usados:

### Scatters

```python
trace = go.Scatter(x=df['sepal_length'], y=df['sepal_width'], mode='markers', marker={'size': 6}, name='sepal', text=df['species'])
fig = go.Figure(data=[trace], layout=go.Layout(xaxis=dict(title= 'longitud', ticklen= 5, zeroline= False),yaxis=dict(title= 'ancho', ticklen= 5, zeroline= False)))
fig.show()
```

### Bar

```python
media_sepalo = df.groupby(['species']).mean().sort_values(by=['sepal_length'])
#creamos lienzo 
fig = go.Figure()
#creamos bucle simple para crear el trazado 
for i in df['species'].unique():
    fig.add_trace(go.Bar(x=media_sepalo.index, y=media_sepalo['sepal_length'], name='species'))
    fig.update_layout(barmode='group', xaxis_tickangle=-45)
    fig.update_traces(marker_line_width = 0.9)
    fig.update_yaxes(title='longitud')
    fig.update_xaxes(tickfont=dict(family='Rockwell', color='gray', size=14),title = '',title_font_family = "Rockwell",title_font_color = "gray",  title_font_size = 20)
    fig.update_layout(legend=dict(x=.1, y=-.2))
    fig.show()
```

### PIE

```python
fig = go.Figure()
fig.add_trace(go.Pie(labels=df['species'], values=df['sepal_length'],name='sepal')) 
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=14, marker=dict(line=dict(color='#000000', width=2)))
fig.update_layout(legend=dict(x=.1, y=-.2))
fig.show()
```

### Heatmap

```python
# Convertimos las especies a números mediante el uso de diccionario 
df['species_numerical'] = df['species'].map({'setosa':0, 'versicolor':1, 'virginica':2})

# Calculamos la correlación entre las columnas seleccionadas y obtenemos la matriz 
df_corr = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species_numerical']].corr()

# Creamos un mapa de calor con la matriz de correlación generada
fig = go.Figure(data=go.Heatmap(z=df_corr.values, x=df_corr.columns, y=df_corr.columns, colorscale='Viridis'))
fig.update_layout(width=500, height=500, title='Mapa de calor')

# Mostramos la figura
fig.show()
```

### Contour

```python
#Creamos Lienzo
fig = go.Figure()
#Creamos trazado 
trace = go.Contour(z=df['petal_length'], x=df['sepal_length'], y=df['sepal_width'])
#Actualizamos layout
fig.update_layout(title="Mapa Contornos", template= "plotly_dark", width=500, height=500)
#Agregamos trazado
fig.add_trace(trace)

#Mostramos figura
fig.show()
```

## Gráficos Financieros , usados para finanzas

```python
# Datos ficticios de ejemplo
import numpy as np
dates = pd.date_range('2020-01-01', '2023-01-31')
open_data = np.random.rand(len(dates))
high_data = open_data + np.random.rand(len(dates))/2
low_data = open_data - np.random.rand(len(dates))/2
close_data = open_data + np.random.rand(len(dates)) - 0.5
```

### OHCL (Precios de apertura, máximo, mínimo y cierre)

```python
#creeamos lienzo
fig = go.Figure()
#generamos trazado
fig.add_trace(go.Ohlc(x=dates, open=open_data, high=high_data, low=low_data, close=close_data,name='Precio bitcoin'))
#actualizamos layout
fig.update_layout(
    title='Precio bitcoin durante tres años',
    yaxis_title='Bitocoin ($) ',
    title_text='Gráfico de Velas',
    template="plotly_dark"
)
#mostramos figura
fig.show()
```

### CandleStick

```python
#creamos lienzo
fig = go.Figure()
#Creamos trazado
fig.add_trace(go.Candlestick(x=dates, open=open_data, high=high_data, low=low_data, close=close_data,name='Precio bitcoin'))
#actualizamos layout   
fig.update_layout(
    title='Precio bitcoin durante tres años',
    yaxis_title='Bitocoin ($) ',
    title_text='Gráfico de Velas',
    template="plotly_dark"
)
#mostramos figura
fig.show()
```

### Waterfall Charts

```python
# Creamos conjuntos de datos ficticios de ejemplo
x_data = ['Ventas', 'Devoluciones', 'Descuentos', 'Fees', 'Ganancia Neta', 'Cheques', 'Créditos', 'Aplicaciones']
y_data = [400, -50, -50, -20, 280, 300, 290, 650]

# Creamos el trazo de cascada
trace = go.Waterfall(
    x=x_data,
    y=y_data,
    measure=['relative']*4 + ['total'] + ['relative']*2 + ['total'],
    base=1000,  # Añadido un valor base para mostrar mejor los incrementos y decrementos
    decreasing={'marker':{'color':'red'}},  # Colores personalizados para incrementos y decrementos
    increasing={'marker':{'color':'green'}},
    totals={'marker':{'color':'blue'}}
)

# Creamos el lienzo
fig = go.Figure(data=[trace])

# Actualizamos el diseño
fig.update_layout(
    template='plotly_dark',
    title_text='Resumen de Operaciones Financieras',
    yaxis_title='Balance en $',
    xaxis={'type': 'category'},  # Esto garantiza que las etiquetas en el eje x sean tratadas como etiquetas, no como números
)

# Mostramos la figura
fig.show()
```

### Funnel Charts

```python
# Creamos el lienzo
fig = go.Figure()

# Definimos los trazados
fig.add_trace(go.Funnel(
    y=["Visitas", "Cliks en producto", "Compran precio", "Solicitan presupuesto", "Compra"],
    x=[39, 27.4, 20.6, 11, 3],
    textinfo="value+percent total",  # Modificado para mostrar porcentaje del total en lugar del inicial
    opacity=0.7,  # Añadido un poco de opacidad
    marker={"color": ["deepskyblue", "lightsalmon", "tan", "teal", "silver"]},  # Añadido colores
    name="Funnel de ventas"
))

# Actualizamos el diseño
fig.update_layout(
    title="Análisis del Funnel de Ventas",
    template="plotly_dark",
    yaxis_title="Etapa del embudo de ventas",
    xaxis_title="Número de usuarios"
)

# Mostramos la figura
fig.show()
```

### Funnel Area Charts

```python
# Creamos el lienzo
fig = go.Figure()

# Definimos el trazado
fig.add_trace(go.Funnelarea(
    labels = ["Visitas", "Cliks en producto", "Compran precio", "Solicitan presupuesto", "Compra"],
    values = [39, 27.4, 20.6, 11, 3],
    textinfo = "value+percent",  # Mostrar valor y porcentaje del total
    marker = dict(colors=['deepskyblue', 'lightsalmon', 'tan', 'teal', 'silver']),  # Colores para cada etapa
))

# Actualizamos el diseño
fig.update_layout(
    title="Análisis del Funnel de Ventas",
    template="plotly_dark",
)

# Mostramos la figura
fig.show()


```

## Malla Tridimensional

```python
i = np.array([0, 0, 0, 1, 1, 1, 2, 2, 2])
j = np.array([1, 2, 3, 0, 2, 3, 0, 1, 3])
k = np.array([2, 3, 1, 2, 3, 0, 1, 3, 0])
x = np.array([0, 1, 2, 0])
y = np.array([0, 0, 1, 2])
z = np.array([0, 2, 0, 1])
```

```python
#creeamos lienzo
fig = go.Figure()
#agregamos trazo
fig.add_trace(go.Mesh3d(x=x, y=y, z=z, i=i, j=j, k=k, name='piramideilluminati'))
#actualizamos layout
fig.update_layout(scene = dict(
        xaxis = dict(nticks=4, range=[-1,2],),
                        yaxis = dict(nticks=4, range=[-1,2],),
                        zaxis = dict(nticks=4, range=[-1,2],),),
                    title="Malla en 3D", template="plotly_dark"
                    )
#Mostramos figura
fig.show()
```

## Superficie 

```python
x = np.outer(np.linspace(-2, 2, 100), np.ones(100))
y = x.copy().T
z = np.cos(x ** 2 + y ** 2)
```

```python
#creeamos lienzo
fig = go.Figure()
#agregamos trazo
fig.add_trace(go.Surface(x=x, y=y, z=z, colorscale='Viridis'))
#actualizamos layout
fig.update_layout(title='Superficie 3D ', autosize=False,width=500, height=500,
                    template="plotly_dark")
#Mostramos figura
fig.show()
```

## Cono

```python
#creeamos lienzo
fig = go.Figure()
#agregamos trazo
fig.add_trace(go.Cone(x=[2, 1, 0], y=[0, 1, 2], z=[0, 2, 0], u=[0, 0, 0], v=[1, -1, 1], w=[1, 1, -1],
                        sizemode="absolute",sizeref=0.5),
                )
#Actualizamos layout
fig.update_layout(title='Cono en 3D', autosize=False,width=500, height=500,
                    template="plotly_dark")
#Mostramos figura
fig.show()
```

## Complementarios

### Scatterpolar

```python
fig = go.Figure(data=go.Scatterpolar(
    r=[2, 5, 8, 4, 6],
    theta=['A', 'B', 'C', 'D', 'E'],
    fill='toself'
))

fig.update_layout(
    polar=dict(radialaxis=dict(visible=True)),
    showlegend=False
)

fig.show()
```

### BarPolar

```python
fig = go.Figure(data=go.Barpolar(
    r=[2, 5, 8, 4, 6],
    theta=['A', 'B', 'C', 'D', 'E'],
    marker_color=["blue", "orange", "green", "red", "purple"],
    marker_line_color="black",
    marker_line_width=2
))

fig.show()
```

### ScatterTernary

```python
fig = go.Figure(data=go.Scatterternary(
    a=[1, 2, 3, 4, 5],
    b=[2, 3, 1, 4, 5],
    c=[3, 1, 2, 5, 4],
    mode='markers',
    marker=dict(size=12, color='Blue')
))

fig.show()
```

### Sunburst

```python
fig = go.Figure(go.Sunburst(
    labels=["A", "B", "C", "D", "E", "F"],
    parents=["", "A", "A", "B", "B", "C"],
    values=[10, 20, 30, 40, 50, 60],
))

fig.update_layout(margin=dict(t=0, l=0, r=0, b=0))

fig.show()
```

### Three Map

```python
fig = go.Figure(go.Treemap(
    labels=["A", "B", "C", "D", "E", "F"],
    parents=["", "A", "A", "B", "B", "C"],
    values=[10, 20, 30, 40, 50, 60],
))

fig.show()
```

### Sankey

```python
fig = go.Figure(data=[go.Sankey(
    node = dict(
        label = ["A", "B", "C", "D"],
    ),
    link = dict(
        source = [0, 1, 0, 2],
        target = [2, 3, 3, 3],
        value = [8, 4, 2, 8]
    ))])

fig.show()
```
